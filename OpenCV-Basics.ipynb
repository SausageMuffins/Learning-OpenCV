```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 
```

```python
img = cv2.imread("C://Users//nryan//OneDrive//Desktop//00-puppy.jpg")

# open cv automatically changes the image to a numpy array.

type(img)
```

```python
# Displaying Images

plt.imshow(img) # open cv and matplotliv have different order in it's channel! -> BGR instead of RGB

# Reverse or Transforming the image.
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fix_img) # fixed img
```

```python
# manipulating images/arrays using cv2

new_img = cv2.resize(fix_img, (1000,400))
plt.imshow(new_img)

# downscaling images
w_ratio = 0.2 # 20 percent of the original width 
h_ratio = 0.2

new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
plt.imshow(new_img)
new_img.shape

# flipping images - useful for generating more images.
new_img = cv2.flip(fix_img, 0) # flip x axist
plt.imshow(new_img)
new_img = cv2.flip(fix_img, 1) # flip y axist
plt.imshow(new_img) 

new_img = cv2.flip(fix_img, -1) # flip both x and y axis
```

```python
# saving images using opencv.

cv2.imwrite("newFile.jpg", fix_img)
```

```python
# designate a larger blank canvas to show the image.
fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111) 
ax.imshow(fix_img)
```

```python
# Placing items on images: important later on to track objects.
blank_img = np.zeros(shape=(512,512,3), dtype=np.int16) # create blank image

# Shapes
pt1 = (200,200)
pt2 = (300,300)

# drawing a rectangle - pt1 and pt2 are the top left and right corner respectively.
cv2.rectangle(blank_img, pt1, pt2, color=(0,255,0), thickness=10)

# drawing a circle
cv2.circle(blank_img, center=(100,100), color=(0,0,255), radius=40, thickness=-1) # thickness = -1 fills the shape.

# drawing a line
cv2.line(blank_img, pt1, pt2, color=(255,0,0), thickness=10)

# drawing a polygon
vertices = np.array([ [50,300], [100,200], [400,350], [200,400]], dtype=np.int32) # establishing vertices
vertices.reshape((-1,1,2)) # reshape to fit the colour channels (this must be done to make the vertices acceptable for cv2)

cv2.polylines(blank_img, [vertices], isClosed=True, color=(255,255,255), thickness = 10)


# writing text on the image.
font = cv2.FONT_HERSHEY_SIMPLEX # assign a font
cv2.putText(blank_img, text="Texted!", org=(10,500), fontFace = font, fontScale = 4, color=(255,255,255), thickness=10)

plt.imshow(blank_img)
```
